## Was ist ein Jupyter-Notebook?

Ein Jupyter-Notebook ist ein webbasiertes Dokument, welches aus Zellen besteht. Es gibt drei Arten von Zellen:
- Markdown-Zellen. Sie enthalten formatierte Texte, Kommentare usw. Die Texte werden in einem Markdown-Format beschrieben. Dieser Text steht in einer Markdown-Zelle.
- Codezellen (Eingabezellen). Sie enthalten Python-Code, der ausgef√ºhrt werden soll
- Ausgabezellen. Sie enthalten die Ausgaben von dem Python-Code in der vorhergehenden Eingabezelle.

Markdown-Zellen und Codezellen haben zwei Modi:
- Editiermodus: Hier kann der Python- bzw. Markdown-Code eingegeben werden

### Titel, und Achsen und Legende mit LaTeX

Strings in Matplotlib wie Titel und Achsenbeschriftungen k√∂nnen **LaTeX-Code** enthalten. Text zwischen `$`-Zeichen wird dabei wie gewohnt im _Math-Mode_ gesetzt.

> **Achtung:** Einige Zeichen wie `\t` (`<TAB>`) oder `\n` (`<RETURN>`) werden von Python als Steuerzeichen und daher nicht als LaTeX-Code interpretiert. Markiert Strings, die LaTeX-Code enthalten, daher mit dem Prefix `r` (f√ºr _raw_), sodass Steuerzeichen ignoriert werden.

````python
x = N.linspace(0, 2 * N.pi, 100)
# Plot mit Label f√ºr Legende
P.plot(x, N.sin(x), label=r'$A \times \sin(\phi)$')
P.plot(x, N.cos(x), label=r'$A \times \cos(\phi)$')
# Titel
P.title('Oszillation')
# Achsenlimits
P.xlim(0, 2 * N.pi)
P.ylim(-1, 1)
# Achsenbeschriftungen
P.xlabel(r'Winkel $\phi \, [\mathrm{rad}]$')
P.ylabel(r'Auslenkung $d \, [\mathrm{cm}]$')
# Legende
P.legend(loc='lower left')
````

## Mehrere Plots in einer Abbildung

Wir k√∂nnen eine Abbildung in ein Gitter mit mehreren Achsen aufteilen. Funktionsaufrufe wie `P.plot` plotten dann immer in die zuletzt erstellte Achse:

````python
# Erstelle eine Abbildung mit Titel
fig = P.figure()
fig.suptitle("Oszillation")

# Erstelle eine Achse in einem Gitter von 2 Zeilen und 1 Spalte
ax_upper = P.subplot(211) # Abk√ºrzung f√ºr die 1. Achse im Gitter

# Funktionen plotten nun in diese Achse:
x = N.linspace(0, 2 * N.pi, 100)
P.plot(x, N.sin(x), label=r'$\sin(\phi)$')
P.xlim(0, 2 * N.pi)
P.legend(loc='lower left')

# Erstelle eine zweite Achse im gleichen Gitter, aber an 2. Position
ax_lower = P.subplot(212, sharex=ax_upper, sharey=ax_upper) # Verwende die gleichen x- und y-Limits wie oben

# Funktionen plotten nun in die zweite Achse:
P.plot(x, N.cos(x), label=r'$\cos(\phi)$')
P.xlabel(r'Winkel $\phi \, [\mathrm{rad}]$')
P.legend(loc='lower left')
````

> **Hinweis:** Matplotlib arbeitet mit _Abbildungen_ und _Achsen_. Wenn wir solche nicht explizit erstellen, wie im Beispiel oben, werden sie beim Aufrufen von Funktionen wie `P.plot` automatisch erstellt. Man kann sie auch jederzeit abrufen:
